<a href="https://tinder.com/" target="_blank">Tinder</a> is an online dating and geosocial networking application. In Tinder, users "swipe right" to like or "swipe left" to dislike other users' profiles, which include their photos, a short bio, and a list of their interests.

## Project üöß

The marketing team needs help on a new project. They are experiencing a decrease in the number of matches, and they are trying to find a way to understand **what makes people interested into each other**. 

They decided to run a speed dating experiment with people who had to give Tinder lots of informations about themselves that could ultimately reflect on ther dating profile on the app.

Tinder then gathered the data from this experiment. Each row in the dataset represents one speed date between two people, and indicates wether each of them secretly agreed to go on a second date with the other person.

## Scope of this project üñºÔ∏è

Data was gathered from participants in experimental speed dating events from 2002-2004. During the events, the attendees would have a four minute "first date" with every other participant of the opposite sex. At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests.

The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information. See the Speed Dating Data Key document below for details.

# Etape 1 : explorer les valeurs manquantes

```python
# Vue d'ensemble des valeurs manquantes
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_info = pd.DataFrame({
    'colonnes': df.columns,
    'nb_manquantes': df.isnull().sum(),
    'pourcentage': missing_percent
}).sort_values('pourcentage', ascending=False)

# Afficher les colonnes avec des donn√©es manquantes
print("=== COLONNES AVEC VALEURS MANQUANTES ===\n")
print(missing_info[missing_info['nb_manquantes'] > 0].head(20))
```

```python
# Cat√©goriser les colonnes selon leur compl√©tude
colonnes_completes = missing_info[missing_info['pourcentage'] == 0]['colonnes'].tolist()
colonnes_utilisables = missing_info[(missing_info['pourcentage'] > 0) & 
                                    (missing_info['pourcentage'] < 30)]['colonnes'].tolist()
colonnes_problematiques = missing_info[missing_info['pourcentage'] >= 30]['colonnes'].tolist()

print(f"=== BILAN ===")
print(f"Colonnes compl√®tes (0% manquant) : {len(colonnes_completes)}")
print(f"Colonnes utilisables (< 30% manquant) : {len(colonnes_utilisables)}")
print(f"Colonnes probl√©matiques (‚â• 30% manquant) : {len(colonnes_problematiques)}")

# Pourquoi ces cat√©gories ?
# Compl√®tes : Id√©ales pour l'analyse
# Utilisables : On peut les garder avec prudence
# Probl√©matiques : √Ä √©viter sauf si vraiment cruciales
```

```python
# Examiner les colonnes essentielles pour le projet
colonnes_critiques = ['dec', 'dec_o', 'match', 'gender', 
                      'attr', 'sinc', 'intel', 'fun', 'amb', 'shar']

print("=== √âTAT DES COLONNES CRITIQUES ===")
for col in colonnes_critiques:
    nb_manquant = df[col].isnull().sum()
    pct_manquant = (nb_manquant / len(df)) * 100
    print(f"{col:10s} : {nb_manquant:5d} manquantes ({pct_manquant:.1f}%)")

# Ces colonnes sont critiques car :
# dec, dec_o, match : Variables cibles (qui dit oui/non)
# gender : Analyse hommes vs femmes
# attr, sinc, etc. : Notes donn√©es pendant le rendez-vous
```

```python
# Visualiser la distribution des valeurs manquantes

# Top 30 des colonnes avec le plus de valeurs manquantes
top_missing = missing_info[missing_info['nb_manquantes'] > 0].head(30).sort_values('pourcentage', ascending=True)

fig = px.bar(
    top_missing,
    y='colonnes',
    x='pourcentage',
    orientation='h',
    title='Top 30 des colonnes avec valeurs manquantes',
    labels={'pourcentage': 'Pourcentage de valeurs manquantes (%)', 'colonnes': 'Colonnes'},
    color='pourcentage',
    color_continuous_scale='Reds'
)

# Ajouter une ligne verticale au seuil de 30%
fig.add_vline(x=30, line_dash="dash", line_color="red", 
              annotation_text="Seuil 30%", annotation_position="top")

fig.update_layout(height=800, showlegend=False)
fig.show()
```

# Etape 2 : Identifier les colonnes cl√©s

Avant de s√©lectionner, r√©capitulons ce qu'on a trouv√©. Bas√© sur le graphique, beaucoup de colonnes du Time 3 (follow-up 3-4 semaines apr√®s) ont ~75-93% de donn√©es manquantes.

```python
# Colonnes essentielles pour l'analyse des matchs
colonnes_cles = [
    'iid', 'gender', 'age', 'race', 'samerace',
    'dec', 'dec_o', 'match',  # D√©cisions
    'attr', 'sinc', 'intel', 'fun', 'amb', 'shar',  # Notes donn√©es
    'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o',  # Notes re√ßues
    'attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1',  # Pr√©f√©rences d√©clar√©es
    'wave', 'order', 'round'
]

# Cr√©er un sous-dataset avec ces colonnes
df_analyse = df[colonnes_cles].copy()
print(f"Donn√©es manquantes dans les colonnes cl√©s :")
print(f"{df_analyse.isnull().sum()}")
```

```python
# D√©finir les colonnes par cat√©gorie d'importance

# COLONNES ESSENTIELLES
colonnes_decision = ['iid', 'id', 'gender', 'wave', 'round', 'order', 'position', 
                     'partner', 'pid', 'dec', 'dec_o', 'match']

colonnes_evaluations = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar',  # Notes donn√©es
                        'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o',  # Notes re√ßues
                        'like', 'prob', 'like_o', 'prob_o']  # Appr√©ciation globale

colonnes_preferences_avant = ['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1',  # Ce que je cherche
                              'attr2_1', 'sinc2_1', 'intel2_1', 'fun2_1', 'amb2_1', 'shar2_1',  # Ce que l'autre sexe cherche
                              'attr3_1', 'sinc3_1', 'intel3_1', 'fun3_1', 'amb3_1',  # Auto-√©valuation
                              'attr4_1', 'sinc4_1', 'intel4_1', 'fun4_1', 'amb4_1', 'shar4_1']  # Ce que mon sexe cherche

colonnes_demographiques = ['age', 'age_o', 'race', 'race_o', 'samerace', 
                          'field_cd', 'career_c', 'income']

colonnes_contexte = ['int_corr',  # Corr√©lation d'int√©r√™ts
                     'exphappy', 'expnum']  # Attentes avant l'√©v√©nement

# Activit√©s/hobbies (pour analyser les int√©r√™ts communs)
colonnes_activites = ['sports', 'tvsports', 'exercise', 'dining', 'museums', 'art', 
                      'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater', 
                      'movies', 'concerts', 'music', 'shopping', 'yoga']

# Cr√©er le dataset avec toutes les colonnes pertinentes
colonnes_finales = (colonnes_decision + colonnes_evaluations + 
                   colonnes_preferences_avant + colonnes_demographiques + 
                   colonnes_contexte + colonnes_activites)

# V√©rifier que toutes les colonnes existent
colonnes_finales = [col for col in colonnes_finales if col in df.columns]

df_clean = df[colonnes_finales].copy()

print(f"Dataset original : {df.shape}")
print(f"Dataset s√©lectionn√© : {df_clean.shape}")
print(f"Colonnes conserv√©es : {len(colonnes_finales)}")

# Explication de la s√©lection :
# On garde les colonnes du Time 1 (avant/pendant l'√©v√©nement) qui ont peu de valeurs manquantes
# On √âLIMINE les colonnes du Time 2 et Time 3 (follow-up apr√®s l'√©v√©nement) qui ont 70-93% de donn√©es manquantes
# On se concentre sur les donn√©es disponibles pour TOUS les participants
```

```python
# V√©rifier l'√©tat des valeurs manquantes dans le dataset nettoy√©
missing_clean = pd.DataFrame({
    'colonnes': df_clean.columns,
    'nb_manquantes': df_clean.isnull().sum(),
    'pourcentage': (df_clean.isnull().sum() / len(df_clean)) * 100
}).sort_values('pourcentage', ascending=False)

print("=== VALEURS MANQUANTES APR√àS S√âLECTION ===")
print(missing_clean[missing_clean['nb_manquantes'] > 0].head(15))
```

```python
# Visualiser l'am√©lioration avec Plotly
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Bar(
    name='Dataset original',
    x=['Colonnes', 'Lignes', 'Valeurs manquantes (%)'],
    y=[df.shape[1], df.shape[0], df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100],
    marker_color='lightcoral'
))

fig.add_trace(go.Bar(
    name='Dataset nettoy√©',
    x=['Colonnes', 'Lignes', 'Valeurs manquantes (%)'],
    y=[df_clean.shape[1], df_clean.shape[0], df_clean.isnull().sum().sum() / (df_clean.shape[0] * df_clean.shape[1]) * 100],
    marker_color='lightgreen'
))

fig.update_layout(
    title='Comparaison : Dataset original vs nettoy√©',
    barmode='group',
    yaxis_title='Valeur'
)

fig.show()
```

# Etape 3 : Analyse descriptive

```python
# Taux de match et d√©cisions
print("="*50)
print("STATISTIQUES GLOBALES SUR LES D√âCISIONS")
print("="*50)

taux_match = df_clean['match'].mean() * 100
taux_dec_oui = df_clean['dec'].mean() * 100
taux_dec_o_oui = df_clean['dec_o'].mean() * 100

print(f"\nTaux de match global : {taux_match:.2f}%")
print(f"Taux de 'oui' des participants : {taux_dec_oui:.2f}%")
print(f"Taux de 'oui' des partenaires : {taux_dec_o_oui:.2f}%")

# Nombre total de rencontres
print(f"Nombre total de rencontres : {len(df_clean)}")
print(f"Nombre de matchs r√©ussis : {df_clean['match'].sum()}")

# Ce qu'on cherche : 
# Le taux de match dira si les gens sont s√©lectifs ou pas
# Un taux de 15-20% signifie que les participants sont exigeants
```

```python
# Analyse par genre
print("\n" + "="*50)
print("COMPARAISON HOMMES vs FEMMES")
print("="*50)

stats_genre = df_clean.groupby('gender').agg({
    'dec': 'mean',
    'attr': 'mean',
    'sinc': 'mean',
    'intel': 'mean',
    'fun': 'mean',
    'amb': 'mean',
    'shar': 'mean'
}).round(2)

stats_genre.columns = ['Taux oui (%)', 'Attractivit√©', 'Sinc√©rit√©', 
                       'Intelligence', 'Fun', 'Ambition', 'Int√©r√™ts partag√©s']
stats_genre['Taux oui (%)'] = stats_genre['Taux oui (%)'] * 100
stats_genre.index = ['Femmes', 'Hommes']

print(stats_genre)

# Ce qu'on analyse :
# Les hommes disent-ils plus souvent "oui" que les femmes ? >> oui (47% vs 37% pour les femmes)
# Quels crit√®res sont not√©s diff√©remment selon le genre ? >> Attractivit√© plus important pour les hommes que pour les femmes, les autres crit√®res sont √©quivalents
```

```python
# Visualisation : Taux de d√©cision par genre

# Pr√©parer les donn√©es
dec_par_genre = df_clean.groupby('gender')['dec'].agg(['mean', 'count']).reset_index()
dec_par_genre['mean'] = dec_par_genre['mean'] * 100
dec_par_genre['gender'] = dec_par_genre['gender'].map({0: 'Femmes', 1: 'Hommes'})

fig = px.bar(
    dec_par_genre,
    x='gender',
    y='mean',
    text='mean',
    title="Taux de d√©cision positive ('oui') par genre",
    labels={'mean': 'Pourcentage de oui (%)', 'gender': 'Genre'},
    color='gender',
    color_discrete_map={'Femmes': '#FF6B9D', 'Hommes': '#4A90E2'}
)

fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.update_layout(showlegend=False, yaxis_range=[0, 100])
fig.show()
```

```python
# Importance des crit√®res : pr√©f√©rences d√©clar√©es vs r√©elles

# Pr√©f√©rences d√©clar√©es (avant l'√©v√©nement)
preferences_declarees = df_clean[['attr1_1', 'sinc1_1', 'intel1_1', 
                                   'fun1_1', 'amb1_1', 'shar1_1']].mean()

# Notes r√©elles donn√©es pendant l'√©v√©nement (corr√©lation avec la d√©cision)
correlations_decision = df_clean[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'dec']].corr()['dec'].drop('dec')

# Cr√©er un DataFrame comparatif
comparaison = pd.DataFrame({
    'Crit√®re': ['Attractivit√©', 'Sinc√©rit√©', 'Intelligence', 'Fun', 'Ambition', 'Int√©r√™ts partag√©s'],
    'Importance d√©clar√©e': preferences_declarees.values,
    'Corr√©lation r√©elle avec d√©cision': correlations_decision.values
}).round(3)

print("\n" + "="*50)
print("PR√âF√âRENCES D√âCLAR√âES vs COMPORTEMENT R√âEL")
print("="*50)
print(comparaison)

# Question cl√© : 
# Les gens disent-ils vouloir quelque chose mais choisissent en fonction d'autre chose ?
# Le top3 des crit√®res avec l'importance d√©clar√©e la plus √©lev√©e sont : Attractivit√©, Intelligence, Fun
# Le top3 des crit√®res avec la corr√©lation r√©elle avec la d√©cision la plus √©lev√©e sont : Attractivit√©, Fun, Int√©r√™ts partag√©s
# Il y a donc un d√©calage entre l'importance d√©clar√©e et l'importance r√©elle
```

```python
# Visualisation : Pr√©f√©rences d√©clar√©es vs r√©elles

fig = go.Figure()

fig.add_trace(go.Bar(
    name='Importance d√©clar√©e',
    x=comparaison['Crit√®re'],
    y=comparaison['Importance d√©clar√©e'],
    marker_color='lightblue'
))

fig.add_trace(go.Scatter(
    name='Corr√©lation r√©elle avec d√©cision',
    x=comparaison['Crit√®re'],
    y=comparaison['Corr√©lation r√©elle avec d√©cision'],
    mode='lines+markers',
    marker=dict(size=10, color='red'),
    line=dict(width=3),
    yaxis='y2'
))

fig.update_layout(
    title='Pr√©f√©rences d√©clar√©es vs Impact r√©el sur la d√©cision',
    xaxis_title='Crit√®res',
    yaxis=dict(title='Importance d√©clar√©e (points)', side='left'),
    yaxis2=dict(title='Corr√©lation avec d√©cision', side='right', overlaying='y'),
    legend=dict(x=0.7, y=1.1),
    hovermode='x unified'
)

fig.show()

# Interpr√©tation : Si l'attractivit√© a une faible importance d√©clar√©e mais une forte corr√©lation r√©elle, cela signifie que les gens sous-estiment son impact r√©el
```

```python
# Moyennes des 6 crit√®res
criteres = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']
noms_criteres = ['Attractivit√©', 'Sinc√©rit√©', 'Intelligence', 'Fun', 'Ambition', 'Int√©r√™ts partag√©s']

moyennes = [df_clean[col].mean() for col in criteres]

fig = px.bar(
    x=noms_criteres,
    y=moyennes,
    title='Moyennes des notes attribu√©es pour chaque crit√®re',
    labels={'x': 'Crit√®res', 'y': 'Note moyenne (sur 10)'},
    text=[f'{m:.2f}' for m in moyennes],
    color=moyennes,
    color_continuous_scale='Blues'
)

fig.update_traces(textposition='outside')
fig.update_layout(showlegend=False, yaxis_range=[0, 10])
fig.show()

# Interpr√©tation : 
# Quel crit√®re re√ßoit les meilleures notes en moyenne ? >> Intelligence
# Lequel est le plus s√©v√®rement not√© ? >> Int√©r√™ts partag√©s
```

```python
# S√©lectionner les crit√®res d'√©valuation
criteres_eval = df_clean[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].dropna()

# Calculer la matrice de corr√©lation
corr_matrix = criteres_eval.corr()

# Cr√©er une heatmap
fig = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=noms_criteres,
    y=noms_criteres,
    colorscale='RdBu',
    zmid=0,
    text=corr_matrix.values.round(2),
    texttemplate='%{text}',
    textfont={"size": 10},
    colorbar=dict(title="Corr√©lation")
))

fig.update_layout(
    title='Corr√©lation entre les crit√®res √©valu√©s',
    width=700,
    height=600
)

fig.show()

# Ce qu'on cherche : 
# Si 2 crit√®res sont tr√®s corr√©l√©s (>0.7), les gens ne les distinguent pas vraiment
# Par exemple, si "fun" et "sinc√©rit√©" sont tr√®s corr√©l√©s, les participants √©valuent ces qualit√©s ensemble
```

```python
# Comparer les notes donn√©es dans les cas de match vs non-match
match_comparison = df_clean.groupby('match')[criteres].mean().T
match_comparison.columns = ['Pas de match', 'Match']
match_comparison['Crit√®re'] = noms_criteres
match_comparison = match_comparison.reset_index(drop=True)

# Visualisation
fig = go.Figure()

fig.add_trace(go.Bar(
    name='Pas de match',
    x=match_comparison['Crit√®re'],
    y=match_comparison['Pas de match'],
    marker_color='lightcoral'
))

fig.add_trace(go.Bar(
    name='Match',
    x=match_comparison['Crit√®re'],
    y=match_comparison['Match'],
    marker_color='lightgreen'
))

fig.update_layout(
    title='Notes moyennes : Match vs Pas de match',
    xaxis_title='Crit√®res',
    yaxis_title='Note moyenne',
    barmode='group',
    yaxis_range=[0, 10]
)

fig.show()

# Question cl√© : 
# Quels crit√®res montrent la plus grande diff√©rence entre match et non-match ? >> Int√©r√™ts partag√©s et Attractivit√©
# Cela r√©v√®le ce qui compte vraiment pour un match mutuel
```

```python
# Comprendre pourquoi certaines sessions ont plus de succ√®s que d'autres

# Cr√©er un tableau r√©capitulatif par session
analyse_waves = df_clean.groupby('wave').agg({
    'match': 'sum',
    'iid': 'count',
    'dec': 'mean',
    'age': 'mean',
    'attr': 'mean',
    'sinc': 'mean',
    'intel': 'mean',
    'fun': 'mean',
    'samerace': 'mean',
    'round': 'first'  # Nombre de personnes rencontr√©es
}).round(2)

analyse_waves.columns = ['Nb matchs', 'Nb rencontres', 'Taux oui', '√Çge moyen', 
                         'Note attr', 'Note sinc', 'Note intel', 'Note fun', 
                         'M√™me origine', 'Nb participants']

analyse_waves['Taux match (%)'] = ((analyse_waves['Nb matchs'] / analyse_waves['Nb rencontres']) * 100).round(1)

print(analyse_waves.sort_values('Taux match (%)', ascending=False))

# Ce qu'on cherche : Corr√©lations entre taux de match et :
# Nombre de participants (trop de monde = fatigue ?)
# √Çge moyen (sessions plus jeunes/√¢g√©es)
# Notes moyennes donn√©es
```

```python
# Sessions avec les taux les plus √©lev√©s et les plus bas
top_sessions = analyse_waves.nlargest(3, 'Taux match (%)')
bottom_sessions = analyse_waves.nsmallest(3, 'Taux match (%)')

print("="*60)
print("TOP 3 SESSIONS (taux de match le plus √©lev√©)")
print("="*60)
print(top_sessions)

print("\n" + "="*60)
print("TOP 3 SESSIONS (taux de match le plus bas)")
print("="*60)
print(bottom_sessions)
```

```python
# Identifier les sessions "succ√®s" (>20%) et "√©chec" (<15%)
df_clean['type_session'] = df_clean['wave'].map(
    lambda x: 'Succ√®s' if analyse_waves.loc[x, 'Taux match (%)'] > 20 
    else ('√âchec' if analyse_waves.loc[x, 'Taux match (%)'] < 15 else 'Moyen')
)

# Comparer les caract√©ristiques
comparaison = df_clean.groupby('type_session').agg({
    'age': 'mean',
    'round': 'mean',
    'attr': 'mean',
    'sinc': 'mean',
    'intel': 'mean',
    'fun': 'mean',
    'amb': 'mean',
    'shar': 'mean',
    'samerace': 'mean',
    'dec': 'mean'
}).round(2)

comparaison.columns = ['√Çge moyen', 'Nb participants', 'Attractivit√©', 'Sinc√©rit√©', 
                       'Intelligence', 'Fun', 'Ambition', 'Int√©r√™ts', 
                       'M√™me origine (%)', 'Taux oui']

print("\n" + "="*60)
print("COMPARAISON : Sessions succ√®s vs √©chec")
print("="*60)
print(comparaison)
```
